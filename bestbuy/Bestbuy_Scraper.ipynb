```python
import pandas as pd
import requests
import re
from bs4 import BeautifulSoup
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
nltk.download('punkt')
nltk.download('wordnet')
```

```python
search_query="intel-core-i5"
base_url="https://www.bestbuy.com/site/searchpage.jsp?st="
url=base_url+search_query
url
header={'User-Agent': 'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/125.0.0.0 Mobile Safari/537.36','Accept-Language': 'da, en-gb, en',
    'Accept-Encoding': 'gzip, deflate, br',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,*/*;q=0.8',
    'Referer': 'https://www.google.com/'}
search_response=requests.get(url,headers=header)
search_response.status_code
search_response.text
search_response.cookies
```

```python
cookie={}
def Searchreviews(review_link):
    url=review_link
    print(url)
    page=requests.get(url,cookies=cookie,headers=header)
    if page.status_code==200:
        return page
    else:
        return "Error"
```

```python
filename = 'bestbuy_urls.txt'
file = open('bestbuy_urls.txt')
```

```python
LIST=re.split("\r\n|\n",file.read())
file.close()
```

```python
url = []
# for i in LIST:
#     print(i.split('~')[-1])
#     url.append(i.split('~')[-1])
url.append(LIST[0].split('~')[-1])
product_name = str(LIST[1])
```

```python
filename.split(".")
```

```python
link = []
pattern = r'(https://www\.bestbuy\.com/site/)([^/]+)(/.+/)(\d+)(\.p\?skuId=)(\d+)'
replacement = r'\1reviews/\2\3\4?\5\6?variant=A'
for i in url:
    modified_url = re.sub(pattern, replacement, i)
    link.append(modified_url)
```

```python
len(link)
print(link)
```

```python
reviews=[]
date=[]
star=[]

# for j in range(len(link)):
for k in range(1,11):
    response=Searchreviews(link[0]+'&page='+str(k))
    soup = BeautifulSoup(response.text, 'lxml')

    for div1 in soup.findAll("div",{'class':"ugc-review-body"}):
        div2 = div1.find('div',{'class':"ugc-components ugc-line-clamp relative"})
        if div2:
            p_tag = div2.find("p",{'class':"pre-white-space"})
            if p_tag:
                print(p_tag.text)
                reviews.append(p_tag.text)
    for div in soup.findAll('div',{'class':"posted-date-ownership disclaimer v-m-right-xxs"}):
        time_tag = div.find("time", {'class': "submission-date"})
        if time_tag:
            title = time_tag.get('title')
            if title:
                print(title)
                date.append(title)
    for n in soup.findAll("p", {'class':"visually-hidden"}):
                    print(n.text)
                    star.append(n.text)
```

```python
star = [stars.replace(' out of 5 stars', '.0') for stars in star]
star = [stars.replace('Rated ', '') for stars in star]
print(star)
```

```python
print(len(date), len(star), len(reviews))
```

```python
len(reviews)
reviews

# Removing Punctuations & Converting to Lowercase
for i in range(len(reviews)):
    reviews[i] = re.sub(r'[^A-Za-z\s]', '', reviews[i]) 
    reviews[i] = reviews[i].lower()


for i in range(len(reviews)):
    reviews[i] = re.sub(r'READ\s*$', '', reviews[i])
    

#Removing Custom Stop Words
custom_words = [
    "against", "above", "below", "out", "under", "again",
    "no", "nor", "not", "don't", "should", "didn't",
    "does", "doesn't", "hadn't", "hasn't", "haven't",
    "isn't", "mustn", "mustn't", "should", "shouldn't",
    "was", "wasn't", "weren", "weren't", "won't", "wouldn't",
    "very", "couldn't"
]

stop_words = set(stopwords.words('english'))
for word in custom_words:
    if word in stop_words:
        stop_words.remove(word)

for i in range(len(reviews)):
    words = re.findall(r'\b\w+\b', reviews[i])
    
    filtered_words = [word for word in words if word.lower() not in stop_words]
    
    filtered_review = ' '.join(filtered_words)
    
    reviews[i] = filtered_review


    

# Lemmetization
lemmatizer = WordNetLemmatizer()
for i in range(len(reviews)):
    words = word_tokenize(reviews[i])
    lemmatized_review = ' '.join([lemmatizer.lemmatize(word) for word in words])
    reviews[i] = lemmatized_review

for review in reviews:
    print(review)
len(review)
len(star)
```

```python
rev = {'Rating': star, 'Date': date, 'Review': reviews}
```

```python
review_data=pd.DataFrame.from_dict(rev)
pd.set_option('max_colwidth',800)
```
