```python
import pandas as pd
import os
import re

names = []
filename = 'merge-filename.txt'
file = open('merge-filename.txt')

LIST=re.split("\r\n|\n",file.read())
file.close()
for i in LIST:
    names.append(i.split('~')[-1])
filename.split(".")

for name in names:
    amazon_path = f'../data/amazon/{name}-a.csv'
    flipkart_path = f'../data/flipkart/{name}-f.csv'
    bestbuy_path = f'../data/bestbuy/{name}-b.csv'

    dataframes = []

    if os.path.exists(amazon_path):
        amazon_df = pd.read_csv(amazon_path)
        amazon_df['Product'] = f'{name}'
        amazon_df['Source'] = 'amazon'
        amazon_df = amazon_df[['Product', 'Rating', 'Date', 'Source', 'Review']]
        dataframes.append(amazon_df)
    else:
        print(f"File not found: {amazon_path}")

    if os.path.exists(flipkart_path):
        flipkart_df = pd.read_csv(flipkart_path)
        flipkart_df['Product'] = f'{name}'
        flipkart_df['Source'] = 'flipkart'
        flipkart_df = flipkart_df[['Product', 'Rating', 'Date', 'Source', 'Review']]
        dataframes.append(flipkart_df)
    else:
        print(f"File not found: {flipkart_path}")

    if os.path.exists(bestbuy_path):
        bestbuy_df = pd.read_csv(bestbuy_path)
        bestbuy_df['Product'] = f'{name}'
        bestbuy_df['Source'] = 'bestbuy'
        bestbuy_df = bestbuy_df[['Product', 'Rating', 'Date', 'Source', 'Review']]
        dataframes.append(bestbuy_df)
    else:
        print(f"File not found: {bestbuy_path}")

    if dataframes:
        merged_df = pd.concat(dataframes, ignore_index=True)
        merged_df.to_csv(f'../data/merged-data/{name}.csv', index=False)
        print("Merged data saved to 'data/merged_reviews.csv'")
    else:
        print("No data to merge.")
```
